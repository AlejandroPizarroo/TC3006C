```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Estadística/Actividades en clase/datasets/breast_cancer.csv')
df.head()
```

```python
df.drop(['id', 'diagnosis'], axis=1, inplace=True)
df.head()
```

```python
correlacion  = df.corr()
alta_corr = np.where((correlacion > 0.95) & (correlacion < 1))
alta_corr
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df_estandar = scaler.fit_transform(df)
df_estandar = pd.DataFrame(df_estandar, columns=df.columns)
df_estandar
```

```python
from sklearn.model_selection import train_test_split
entrenamiento, prueba = train_test_split(df_estandar, test_size=0.20, random_state=42)
entrenamiento
```

```python
entrenamiento.columns
```

```python
import statsmodels.formula.api as smf
modelo = smf.ols(formula='radius_mean~texture_mean+perimeter_mean+area_mean+smoothness_mean+compactness_mean+concavity_mean+symmetry_mean+fractal_dimension_mean+radius_se+texture_se', data = entrenamiento)
modelo = modelo.fit()
print(modelo.summary())
```
